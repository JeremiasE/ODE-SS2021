# Das Lsungsportrt von $\dot x(t)=\frac{1}{tx(t)}$
Beispiel 2.5.4

```python
def plot_solution(x0,t0,color):
    f = sqrt(x0^2+2*log(t/t0))
    if x0<0:
        f=-f
    if t0<0:
        return plot(f,(t,-10,exp(-x0^2/2)*t0),color=color)
    return plot(f,(t,exp(-x0^2/2)*t0,10),color=color)

sf = plot_slope_field(1/(t*x),(t,-3,3),(x,-3,3),color=(0.2,0.2,0.2,0.5))
cp = (sf 
 +plot_solution(1,2,color=colors.firebrick) 
 +plot_solution(-1,-2,color=colors.orange) 
 +plot_solution(-1,2,color=colors.royalblue) 
 +plot_solution(1,-1,color=colors.green)
)
cp.axes_labels(("$t$","$x$"))
cp.save("slope-field-1tx.pdf",xmin=-3,xmax=3)
cp
```
