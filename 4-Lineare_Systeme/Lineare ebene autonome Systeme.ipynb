# Lineare ebene autonome Systeme

```python
%display latex
```

```python
var("t,x,y")
```

```python
def plot_linear_system(A,color="#267ab2",show_vars=True):
    VektorFeld = tuple(A*vector([x,y]))
    Anfangsbedingung = [0,2.2,0]
    Endzeitpunkt = 9

    FieldPlot = plot_vector_field(vector(VektorFeld).normalized(),(x,-8,8),(y,-8,8),color=(0.4,0.4,0.4,0.6))
    StreamPlot = streamline_plot(VektorFeld, (x,-8,8), (y,-8,8),color=color)
    MainPlot =  StreamPlot + FieldPlot
    MainPlot.axes_range(-7,7,-7,7)
    # MainPlot.save_image("pendel-streamplot.pdf",axes_labels=['$x$', '$\\dot x$'])
    if show_vars:
        show(LatexExpr("A="),A)
        show(LatexExpr("e^{tA}="),(exp(t*A)))
        show(LatexExpr("\\sigma(A)="),tuple(A.eigenvalues()))
    MainPlot.show(axes_labels=['$x$', '$y$'],figsize=[5,5])
```

Im Folgenden stellen wir stets Phasenkurven bzw. Trajektorien (also Teilmengen der Ebene, die durch LÃ¶sungen der DGL parametrisiert sind) der DGL 
$$\begin{pmatrix} \dot x(t) \\ \dot y(t) \end{pmatrix} = A \begin{pmatrix} x(t) \\ y(t) \end{pmatrix}$$ dar. 

## Instabiler Knoten

```python
A = matrix([[1,0],[0,1]])
plot_linear_system(A)
```

```python
A = matrix([[2,0],[0,1]])
plot_linear_system(A)
```

```python
A = matrix([[1,1],[0,1]])
plot_linear_system(A)
```

## Stabiler Knoten

```python
A = matrix([[-2,0],[0,-1]])
plot_linear_system(A)
```

```python
A = matrix([[-1,1],[0,-1]])
plot_linear_system(A)
```

## Sattel

```python
A = matrix([[1,0],[0,-1]])
plot_linear_system(A)
```

```python
C = matrix([[1,0],[0,-1]])
P = matrix([[1,0],[1,1]])
plot_linear_system(P*C*P^(-1))
```

## Stabiler Strudel

```python
A = matrix([[-2,-1],[1,-2]])
plot_linear_system(A,color=(0.7,0.1,0.1))
```

```python
A = matrix([[2,-1],[1,2]])
plot_linear_system(A,color=(0.7,0.1,0.1))
```
